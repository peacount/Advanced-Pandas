```python
import os
os.chdir('C:\\Users\\jenny\\venvs\\py39\\Lib\\site-packages')
import pandas as pd
```

```python
names = pd.Series(['Pomeray, CODY ',' Wagner; Jarry','smith, Ray'])
```

```python
names = names.str.replace(';',',')
names
```

```python
names.str.len()
```

```python
names = names.str.strip()
names.str.len()
```

```python
names = names.str.lower()
names

#try .str.upper()
```

```python
names = names.str.split(',')
names
```

```python
names = pd.Series([i[::-1] for i in names])
names
```

```python
names = [' '.join(i) for i in names] 
names
```

The join() method takes all items in an iterable and joins them into one string.

A string must be specified as the separator.

string.join(iterable)

```python
myTuple = ('John', 'Peter', 'Vicky')
x = '#'.join(myTuple)
x
```

```python
myDict = {'name': 'John', 'country': 'Norway'}
mySeperator = 'Test'
x = mySeperator.join(myDict) # When using a dictionary as an iterable, the returned values are the keys, not the values.
x
```
